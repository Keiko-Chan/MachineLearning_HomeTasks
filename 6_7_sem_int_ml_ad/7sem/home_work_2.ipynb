```python
def learn(xtrain, ytrain, xtest, ytest):
    m = LinearRegression()
    m.fit(xtrain, ytrain)
    
    predict = m.predict(xtest)
    mse = mean_squared_error(ytest, predict)
    
    return mse

def del_priznak(x_train, x_test, ytrain, ytest, trash_hold):
    for i in x_train:
        new_x_train = x_train.drop(columns = i)
        new_x_test = x_test.drop(columns = i)
        
        mse = learn(new_x_train, ytrain, new_x_test, ytest)
        if mse < trash_hold:
            print(i, ": ", mse)
    
#del_priznak(X_train, X_test, Y_train, Y_test, 3223699)  
```
